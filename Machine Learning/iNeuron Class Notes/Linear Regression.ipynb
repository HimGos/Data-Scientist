# Training the Model

```python
# Independent & Dependent Features


# Independent Feature. Notate it with X.
# Picking all independent features
X = dataset.iloc[:,:-1]

# Dependent Feature. Notating it with Y.
# Price is our only Dependent Feature.
y = dataset.iloc[:,-1]
```

```python
X.head()
```

```python
y.head()            # We get a series in a Dependent Feature or 1D array.
```

> ###  **Splitting the data into TRAINING & TEST Dataset**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

# Random state is important
# We will use X_test & y_test after the model gets trained
```

```python
# Size of X_train will be equal to y_train
```

```python
X_train.shape
```

```python
y_train.shape
```

> ## In linear regression we get Gradient Descent, with the help of Feature Scaling we can get to global minima quickly

Bringing both features in same unit bring Gradient Descent easily. Eg. If there is one feature in KM and other in Meter, so we match their units to get work done faster.

FEATURE SCALING IS REQUIRED ONLY IN LINEAR REGRESSION. DON't WASTE TIME in upcoming algorithms like XGboost etc.

```python
## Standardize or feature scaling the Dataset

from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
# Standardizing our training data

X_train = scaler.fit_transform(X_train)
```

```python
# We wrote only .transform() to avoid data leakage. We don't want any relation between train and test. Eg. During exam, our X_train is syllabus and X_test is our Exam Paper.
# We don't want to leak Exam paper so we use .transform()
# Check Krish YT Video on Data Leakage

X_test = scaler.transform(X_test)
```

## Model Training

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

```python
# fit() means training the data., trying to find the parameter ..... fit_transform() means send the data and train the data.

regression.fit(X_train, y_train)
```

```python
## Print the coefficients (13) and the intercept

print(regression.coef_)
```

```python
print(regression.intercept_)
```

> ## If all other features are zero then Price of house will be 22!

```python
## Prediction for the test data

reg_pred = regression.predict(X_test)
```

```python
reg_pred[:5]
```

## Assumptions of Linear Regressions

Doing this to check whether our model is good or not

```python
# Comparing actual and predicted points

plt.scatter(y_test,reg_pred)
plt.xlabel("Test Truth Data")
plt.ylabel("Test Predicted Data")
```

> ## We are getting a linear relationship, it means our model is GOOOD!!

```python
## Residuals

residuals = y_test - reg_pred
```

```python
# These are errors

residuals[:5]
```

# Performance Metrics

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_squared_error(y_test, reg_pred))
print(mean_absolute_error(y_test, reg_pred))
print(np.sqrt(mean_squared_error(y_test, reg_pred)))
```

```python
# Computing R sqaure and Adjusted R square

from sklearn.metrics import r2_score

score = r2_score(y_test, reg_pred)
print(score)
```

```python
# Adjusted R square

1 - (1-score)*(len(y_test)-1) / (len(y_test)-X_test.shape[1] - 1)
```
